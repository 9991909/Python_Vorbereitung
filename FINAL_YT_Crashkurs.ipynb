**While Schleife**

-> falls Code so wäre...

while 5 < 10:
   print ("blabla") 

...würde Programm immer wieder Aussage 5<10 prüfen und da dies immer stimmt, würde einfach immer blabla geprintet werden. In diesem Fall müsste man noch einen counter festlegen und 5 damit austauschen + counter so "festlegen", dass es auch über 5 irgendwann liegt



```python
counter = 17
while counter <25:
  print ("blabla")
  counter = counter + 1

#Bedingung wird so oft geprüft, bis counter + 1 nicht mehr kleiner als 25 ist 
```

**Funktionen mit Parametern**

 

1.   bei oberen Funktion sind *keine Parameter* vorhanden, da die *Klammer leer* ist
2.   wenn man ohne Parameter arbeitet, müsste man für jeden Namen der begrüßt werden soll eine eigene Funktion basteln

```python
def say_hello(first_name, last_name): 
  print("Hello " + first_name + " " + last_name + " " + "Whaaaas poppin?") 
  print ("Gang Gang")


say_hello("Ali" ,"Woy")

say_hello("Liselotte","Musterfrau")

print(say_hello("Ali" ,"Woy")) #hier mit Rückgabewert


#in Klammern kommt die beliebige Bezeichnung des Parameters
#Parameter muss dann wie eine Variable eingebaut/verkettet werden
#bis dahin wurde aber wieder nur der Aufbauplan der Funktion erstellt 
#mit alleiniger ausführung der Funktion inklusive eingesetzter Variable findet dann Ausgabe statt 
```

**Funktionen mit Rückgabewert**



1.   *jede Funktion* gibt immer einen *Rückgabewert* zurück 
2.   *None* steht für die *Abwesenheit* eines Wertes -> falls man mit return keinen Wert zurückgegeben will, wird immer None ausgegeben
3.   um sich einen Wert bewusst *zurückgeben lassen* möchte, verwendet man *return*

```python
def maximum(a,b):
  if a < b:
    return b 
  else:
    return a

maximum(3,7) 

result = maximum (7,3)

print(result)

#Funktion soll das Maximum zurückgeben #Parameter
#If-Statement gefolgt von Rückgabe durch return 
#else falls b nicht größter Wert ist 
#Funktion wird ausgeführt und größter Wert wird zurückgegeben
#Ergebnis wird Result zugeordnet, sodass man mit print die Ausgabe des Wertes startet
```

**Konzept der objektorientierten Programmierung**


1.   hilft gewisse Dinge im Code besser darzustellen/*modellieren*
2.   Welt besteht aus *komplexen* Objekten mit *Eigenschaften* -> Basisdatentypen reichen hier nicht aus um Auto bspw. komplett zu repräsentieren
3.   Idee der OOP: Aus *welchen Objekten* setz sich *Programm* zusammen, Objekte *modellieren *(eigene Datentypen bauen), mit restlicher Programmlogik modellierte Objekte *in Beziehung setzen* 
4. Funktionen innerhalb einer Klasse werden als *Methode* bezeichnet
5. nahezu alle Methoden haben den *Self-Parameter*

```python
class Auto:
  def __init__(self):
    self.Auto_Hersteller = None
    self.Auto_PS = None 
    self.Auto_Farbe = None 

Auto1 = Auto()
print(Auto1.Auto_Hersteller)
Auto1.Auto_Hersteller = "Lambo"
Auto1.Auto_PS = 1000
Auto1.Auto_Farbe = "Lavendel"
print(Auto1.Auto_Hersteller)
print(Auto1.Auto_PS)
print(Auto1.Auto_Farbe)




#erst wird Klasse bezeichnet #dann werden durch init Methode bzw. danach Eigenschaften der Methode festgelegt # einzelne Eigenschaften werden mit self formuliert
#init Methode wird immer genau so definiert #Klasse führt ähnlich wie Funktion nicht direkt zu einer Ausgabe, sondern gibt Objekt einen Aufbauplan 
#Klasse ist übergreifender Aufbauplan für die dann konkrete Objekte erzeugt werden 
```
